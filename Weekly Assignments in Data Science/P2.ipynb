```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import mutual_info_classif, f_classif, SelectKBest
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# Load your dataset (assuming it's stored in a DataFrame called 'df')
# Replace this with your actual dataset loading code if not already loaded

# Separate features (X) and target (y)
X = d_data.drop('Diabetes_012', axis=1)
y = d_data['Diabetes_012']

# Identify categorical and continuous features
categorical_features = X.columns[X.dtypes == 'float64'].tolist()
continuous_features = X.columns[X.dtypes != 'float64'].tolist()

# Convert categorical features to 'category' type for mutual information calculation
X[categorical_features] = X[categorical_features].astype('category')

# Use mutual information and ANOVA F-value for feature selection
# Mutual information for both categorical and continuous target variable
mi = mutual_info_classif(X, y, discrete_features='auto', random_state=42)

# ANOVA F-value for categorical features with continuous target variable
anova_f = f_classif(X[categorical_features], y)[0]

# Combine mutual information and ANOVA F-value scores
feature_scores = pd.DataFrame({
    'Feature': X.columns,
    'MI_Score': mi,
    'ANOVA_F': anova_f
})

# Select features based on scores (e.g., top 10 based on highest MI score)
top_features_mi = feature_scores.sort_values(by='MI_Score', ascending=False).head(10)['Feature'].tolist()

# Create a new DataFrame with selected features
X_selected = X[top_features_mi]

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X_selected, y, test_size=0.2, random_state=42)

# Train a base model using all features
base_model = RandomForestClassifier(random_state=42)
base_model.fit(X_train, y_train)
base_preds = base_model.predict(X_test)
base_accuracy = accuracy_score(y_test, base_preds)

# Train a model using selected features
selected_model = RandomForestClassifier(random_state=42)
selected_model.fit(X_train, y_train)
selected_preds = selected_model.predict(X_test)
selected_accuracy = accuracy_score(y_test, selected_preds)

print(f"Base model accuracy: {base_accuracy:.4f}")
print(f"Selected model accuracy: {selected_accuracy:.4f}")

# Output selected features
print("Selected features:")
print(X_selected.columns.tolist())
```

```python
# Create a validation set
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=10)

# Setting up the KNN model with a reduced initial set of neighbors
estimator_KNN = KNeighborsClassifier(n_neighbors=3)
estimator_KNN.fit(X_train, y_train)

# Instantiating the Multinomial Logistic Regression model
log_reg = LogisticRegression(solver='lbfgs', max_iter=100, random_state=10)
log_reg.fit(X_train, y_train)

# Predicting on the test set
y_pred_log_reg = log_reg.predict(X_test)

# Defining the parameters for grid search with a smaller grid
parameters_KNN = {
    'n_neighbors': range(2, 4),
    'leaf_size': range(10, 20),
    'weights': ['uniform', 'distance']
}

# Setting up GridSearchCV with fewer folds for faster evaluation
grid_search_KNN = GridSearchCV(
    estimator=estimator_KNN,
    param_grid=parameters_KNN,
    scoring='accuracy',
    n_jobs=-1,
    cv=3  # Reduce the number of folds
)

# Fitting the grid search to the validation data
grid_search_KNN.fit(X_train, y_train)

# Evaluating and displaying results
print('Accuracy of the initial model:', round(estimator_KNN.score(X_test, y_test),2))
print("Best Parameters:", grid_search_KNN.best_params_)
print("Best Cross-Validation Score:", round(grid_search_KNN.best_score_, 2))

# Getting the best model from the grid search
best_model_KNN = grid_search_KNN.best_estimator_
y_pred = best_model_KNN.predict(X_test)
print("Test Set Accuracy:", round(accuracy_score(y_test, y_pred), 2))
```

SVC

```python
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

# Splitting training data to create a validation subset
X_train_sub, X_val, y_train_sub, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=10)

# Instantiating and fitting the Logistic Regression model
log_reg = LogisticRegression(solver='lbfgs', max_iter=100, random_state=10)
log_reg.fit(X_train_sub, y_train_sub)

# Defining a reduced hyperparameter grid for SVM
param_grid = {
    'C': [0.1, 1],  # Simplified range for C
    'kernel': ['linear', 'rbf'],  
    'gamma': ['scale']  # Using default gamma
}

# Initializing GridSearchCV
grid_search = GridSearchCV(
    estimator=SVC(),
    param_grid=param_grid,
    cv=3,  # Reduced cross-validation folds
    n_jobs=-1,  # Utilize all CPU cores
    scoring='accuracy'
)

# Fit the model using GridSearchCV on the training subset
grid_search.fit(X_train_sub, y_train_sub)

# Retrieving the best model
best_svm_model = grid_search.best_estimator_

# Predicting on the test set with the best SVM model
y_pred_svm = best_svm_model.predict(X_test)

# Calculating accuracy on the test set
svm_accuracy = accuracy_score(y_test, y_pred_svm)

# Printing the results
print("Best Parameters:", grid_search.best_params_)
print("Best Cross-Validation Score:", round(grid_search.best_score_, 2))
print("Test Set Accuracy:", round(svm_accuracy, 2))
```

```python
# Splitting the dataset into a smaller training subset and validation set
X_train_sub, X_val, y_train_sub, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=10)

# Defining the Logistic Regression model
log_reg = LogisticRegression(solver='lbfgs', max_iter=200, random_state=10)

# Defining a simplified hyperparameter grid
param_grid = {
    'penalty': ['l2'], 
    'C': [0.1, 1, 10], 
    'max_iter': [100, 200]  
}

# Initializing GridSearchCV with fewer cross-validation folds
grid_search_LR = GridSearchCV(
    estimator=log_reg,
    param_grid=param_grid,
    cv=3,  
    n_jobs=-1,  
    scoring='accuracy'
)

# Fitting the model using GridSearchCV
grid_search_LR.fit(X_train_sub, y_train_sub)

# Retrieving the best model
best_log_reg = grid_search_LR.best_estimator_

# Predicting on the test set
y_pred = best_log_reg.predict(X_test)

# Evaluating and displaying results
print('Accuracy of the initial model:', grid_search_LR.score(X_test, y_test))
print("Best Parameters:", grid_search_LR.best_params_)
print("Best Cross-Validation Score:", round(grid_search_LR.best_score_, 2))

# Getting the best model from the grid search
best_model_lr = grid_search_LR.best_estimator_
y_pred = best_model_lr.predict(X_test)
print("Test Set Accuracy:", round(accuracy_score(y_test, y_pred), 2))
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score

# Hyperparameters for Grid Search
param_grid_knn = {
    'n_neighbors': [3, 5, 7],
    'weights': ['uniform', 'distance']
}

param_grid_svc = {
    'C': [0.1, 1, 10],
    'kernel': ['linear', 'rbf'],
    'gamma': ['scale', 'auto']
}

param_grid_dt = {
    'max_depth': [5, 10, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

# Models
knn = KNeighborsClassifier()
svc = SVC()
dt = DecisionTreeClassifier()

# GridSearchCV
grid_search_knn = GridSearchCV(knn, param_grid_knn, cv=3, n_jobs=-1, scoring='accuracy')
grid_search_svc = GridSearchCV(svc, param_grid_svc, cv=3, n_jobs=-1, scoring='accuracy')
grid_search_dt = GridSearchCV(dt, param_grid_dt, cv=3, n_jobs=-1, scoring='accuracy')

# Fitting the models
grid_search_knn.fit(X_train, y_train)
grid_search_svc.fit(X_train, y_train)
grid_search_dt.fit(X_train, y_train)

# Best models
best_knn = grid_search_knn.best_estimator_
best_svc = grid_search_svc.best_estimator_
best_dt = grid_search_dt.best_estimator_

# Predicting on the test set
y_pred_knn = best_knn.predict(X_test)
y_pred_svc = best_svc.predict(X_test)
y_pred_dt = best_dt.predict(X_test)

# Evaluating and displaying results
print("KNN Test Set Accuracy:", round(accuracy_score(y_test, y_pred_knn), 2))
print("SVC Test Set Accuracy:", round(accuracy_score(y_test, y_pred_svc), 2))
print("Decision Tree Test Set Accuracy:", round(accuracy_score(y_test, y_pred_dt), 2))
```
